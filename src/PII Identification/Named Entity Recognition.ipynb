```python
import nltk
from nltk.corpus import stopwords
import boto3
from io import StringIO
```

```python
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
session = boto3.Session(profile_name='ai-dataset-bucket-readwrite-795524854110')

# Connect to S3
s3 = session.client('s3')

bucket_name = 'artificial-intelligence-datasets'
object_key = 'input/Text Classification E-mails dataset/Science/14147.txt'  # e.g., 'project1/data/sample.csv'

# Download object into memory
response = s3.get_object(Bucket=bucket_name, Key=object_key)

with response['Body'] as stream:
    text = stream.read().decode('utf-8')
```

```python
words = nltk.word_tokenize(text)
print(f"Number of words = {len(words)}")
print(words)
```
